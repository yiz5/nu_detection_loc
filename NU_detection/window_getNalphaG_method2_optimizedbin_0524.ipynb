```python
etaijk_pd = pd.read_csv('atmhepdsnb_Xe_ETAijk_optimized_consecutive_thrd0_S1S2binning.csv', index_col = 0)
etaijk_pd
```

```python
signal_pcles  =['atm', 'hep', 'dsnb']
```

```python
detectors = ['CJPL', 'Kamioka', 'LNGS', 'SURF', 'SNOlab']
alphas = [0.001, 0.005, 0.01,  0.05 , 0.1]
metallicities = ['high']#, 'low'
```

```python
N_alpha_G = pd.DataFrame(columns=[signal_pcle+'_'+detector+'_'+metallicity  + '_alpha'+str(alpha) + '_'+quantity 
                                  for signal_pcle in signal_pcles
                                  for detector in detectors 
                                  for metallicity in metallicities
                                  for alpha in alphas
                                  for quantity in ['N_alpha', 'G', 'mean']])
N_alpha_G
```

```python
for signal_pcle in signal_pcles:
    if signal_pcle == 'dsnb':
        exposures = [int(i) for i in np.logspace(np.log10(100), np.log10(10000000), 26)]

    elif signal_pcle == 'hep':
        exposures = [int(i) for i in np.logspace(np.log10(10), np.log10(1000000), 26)]

    elif signal_pcle == 'atm':
        exposures = [int(i) for i in np.logspace(np.log10(100), np.log10(10000000), 26)]
    N_alpha_G[signal_pcle+'_'+'exposures'] = exposures
    
    
    for detector in detectors:
    

        for m, metallicity in enumerate(metallicities):

            null_sum_ijkappa = etaijk_pd['null_sum'][signal_pcle + '_'+detector+'_'+metallicity]
            alter_sum_ijkappa = etaijk_pd['alter_sum'][signal_pcle + '_'+detector+'_'+metallicity]


            print('null_sum_ij = ', null_sum_ijkappa, 
                  '\nalter_sum_ij = ', alter_sum_ijkappa)
            
            for alpha in alphas:

                DT, Ns_alpha, powers, Gs, test_exposures = get_nume_DT_alphabeta_G_method2(
                    exposures, null_sum_ijkappa, alter_sum_ijkappa, alpha, 10, 
                        print_check = False, plot = False)
                
                E_Ns_nulls = null_sum_ijkappa * test_exposures
                N_alpha_G[signal_pcle+'_'+detector+'_'+metallicity  + '_alpha'+str(alpha) + '_mean'] = E_Ns_nulls
                N_alpha_G[signal_pcle+'_'+detector+'_'+metallicity  + '_alpha'+str(alpha) + '_N_alpha'] = Ns_alpha
                N_alpha_G[signal_pcle+'_'+detector+'_'+metallicity  + '_alpha'+str(alpha) + '_G'] = Gs
                print(detector, 'alpha = ', alpha , Ns_alpha, Gs)


```
