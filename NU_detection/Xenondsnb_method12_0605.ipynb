```python
def find_consecutive(data, stepsize=1):
    return np.split(data, np.where(np.diff(data) != stepsize)[0]+1)
```

```python

exposures = [int(i) for i in np.logspace(np.log10(1e3), np.log10(1e6), 31)]
Er_bincenters = list(binned_1d_pd['Xenon_Er_bincenters [keV]'])
for eff_corr in effs_corrs:
    eff, corr = eff_corr 
    print(eff_corr)
    for detector in detectors:
        print(detector)
        for metallicity in metallicities:
            pcles_det = ['dsnb']
            select_col_det  = [col for col in binned_1d_pd.columns 
                               if '_'+eff+'_' in col and corr in col and 
                               metallicity in col and any(pcle in col for pcle in pcles_det)]

            pcles_null_NR_woatm =['hep', '8B']
            pcles_null_NR = ['hep', '8B','atmNu_'+detector+'_avg']
            pcles_null_atm_NR = ['atmNu_'+detector+'_avg']
            select_col_null_woatm  = [col for col in binned_1d_pd.columns 
                                      if '_'+eff+'_' in col and corr in col and 
                                      metallicity in col and any(pcle in col for pcle in pcles_null_NR_woatm)]
            select_col_null_atm  = [col for col in binned_1d_pd.columns 
                                    if '_'+eff+'_' in col and corr in col and 
                                    metallicity in col and any(pcle in col for pcle in pcles_null_atm_NR)]

            select_col_null  = [col for col in binned_1d_pd.columns 
                                if '_'+eff+'_' in col and corr in col and 
                                metallicity in col and any(pcle in col for pcle in pcles_null_NR)]

            if len(
                select_col_det) !=len(pcles_det) or len(
                select_col_null) !=len(pcles_null_NR) or len(
                select_col_null_woatm) !=len(pcles_null_NR_woatm) or len(
                select_col_null_atm) !=len(pcles_null_atm_NR):
                print('wrong selection',
                     select_col_det, pcles_det,'\n',
                     select_col_null, pcles_null_NR, '\n',
                     select_col_null_woatm, pcles_null_NR_woatm, '\n',
                     select_col_null_atm, pcles_null_atm_NR)
                break
            else:

                rates_det = list(np.sum(binned_1d_pd[select_col_det], axis = 1))/u.tonne/u.yr
                rates_null = list(np.sum(binned_1d_pd[select_col_null], axis = 1))/u.tonne/u.yr
                rates_null_woatm = list(np.sum(binned_1d_pd[select_col_null_woatm], axis = 1))/u.tonne/u.yr
                rates_null_atm = list(np.sum(binned_1d_pd[select_col_null_atm], axis = 1))/u.tonne/u.yr


                #keep all dsnb
                good1 = rates_det>0

                #remove all bgs except atm and dsnb exists
                good2 = np.multiply(rates_det>0, rates_null_woatm==0)

                #keep dsnb greater than atm
                good3 = rates_null_atm<rates_det 

                #keep dsnb greater than else bgs and greater than fraction of atm
                good4 = (rates_null_atm<4*rates_det) * (rates_null_woatm<rates_det)
                #np.multiply(rates_null_atm<4*rates_det, rates_null_woatm<rates_det)


                start_index1, end_index1 = min(np.where(good1)[0]), max(np.where(good1)[0])
                start_index2, end_index2 = min(np.where(good2)[0]), max(np.where(good2)[0])
                start_index3, end_index3 = min(np.where(good3)[0]), max(np.where(good3)[0])
                start_index4, end_index4 = min([i for i in range(0, len(good4)-1) if good4[i] == True and good4[i-1]==False]), max([i for i in range(0, len(good4)-1) if good4[i] == True and good4[i-1]==True])

                rates_null_good1 = rates_null[start_index1:end_index1+1]
                rates_null_good2 = rates_null[start_index2:end_index2+1]
                rates_null_good3 = rates_null[start_index3:end_index3+1]
                rates_null_good4 = rates_null[start_index4:end_index4+1]

                rates_det_good1 = rates_det[start_index1:end_index1+1]
                rates_det_good2 = rates_det[start_index2:end_index2+1]
                rates_det_good3 = rates_det[start_index3:end_index3+1]
                rates_det_good4 = rates_det[start_index4:end_index4+1]

                null1_sum_ijkappa = sum(rates_null_good1)
                alter1_sum_ijkappa = sum(rates_det_good1 + rates_null_good1)

                null2_sum_ijkappa = sum(rates_null_good2)
                alter2_sum_ijkappa = sum(rates_det_good2 + rates_null_good2)

                null3_sum_ijkappa = sum(rates_null_good3)
                alter3_sum_ijkappa = sum(rates_det_good3 + rates_null_good3)

                null4_sum_ijkappa = sum(rates_null_good4)
                alter4_sum_ijkappa = sum(rates_det_good4 + rates_null_good4)

                print('\nEr range keep all dsnb: ', len(rates_null_good1), 'bins from ', 
                      Er_bincenters[start_index1], ' to ', Er_bincenters[end_index1], 
                      '\nnull_sum_ijkappa = ', null1_sum_ijkappa, 
                      '\nalter_sum_ijkappa = ', alter1_sum_ijkappa,

                     '\n\nEr range remove all bgs except atm and dsnb exists: ', len(rates_null_good2), 'bins from ', 
                      Er_bincenters[start_index2], ' to ', Er_bincenters[end_index2], 
                      '\nnull_sum_ijkappa = ', null2_sum_ijkappa, 
                      '\nalter_sum_ijkappa = ', alter2_sum_ijkappa, 

                      '\n\nEr range keep dsnb greater than atm: ', len(rates_null_good3), 'bins from ', 
                      Er_bincenters[start_index3], ' to ', Er_bincenters[end_index3], 
                      '\nnull_sum_ijkappa = ', null3_sum_ijkappa, 
                      '\nalter_sum_ijkappa = ', alter3_sum_ijkappa, 

                      '\n\nEr range keep dsnb greater than else bgs and greater than fraction of atm: ', 
                      len(rates_null_good4), 'bins from ', 
                      Er_bincenters[start_index4], ' to ', Er_bincenters[end_index4], 
                      '\nnull_sum_ijkappa = ', null4_sum_ijkappa, 
                      '\nalter_sum_ijkappa = ', alter4_sum_ijkappa, 

                      '\n')

                for alpha_power in alphas_powers:
                    dict_key = str(alpha_power)
                    alpha, power = alpha_power
                    Z_alpha = round(stats.norm.ppf(alpha), 3)
                    Phi_minus1 = round(stats.norm.ppf(power), 3)

                    
                    #keep all dsnb
                    DT_method2_good1,_,_,_,_ = get_nume_DT_alphabeta_G_method2(exposures, 
                                        null1_sum_ijkappa.value, alter1_sum_ijkappa.value, 
                                        alpha, power, simnum = 500000, 
                                         plot= False, print_check = False)
                    result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_allDSNBMethod2'][str(alpha_power)] = DT_method2_good1
                    print('keep all dsnb method2 DT = ', DT_method2_good1, '\n')

                    #remove all bgs except atm and dsnb exists
                    DT_method2_good2,_,_,_,_ = get_nume_DT_alphabeta_G_method2(exposures, 
                                        null2_sum_ijkappa.value, alter2_sum_ijkappa.value, 
                                        alpha, power, simnum = 500000, 
                                         plot= False, print_check = False)
                    result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_noelsebgMethod2'][str(alpha_power)] = DT_method2_good2
                    print('remove all bgs except atm and dsnb exists method2 DT = ', DT_method2_good2, '\n')

                    #keep dsnb greater than atm 
                    DT_method2_good3,_,_,_,_ = get_nume_DT_alphabeta_G_method2(exposures, 
                                        null3_sum_ijkappa.value, alter3_sum_ijkappa.value, 
                                        alpha,power, simnum = 500000, 
                                         plot= False, print_check = False)
                    result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_lessatmbgMethod2'][str(alpha_power)] = DT_method2_good3

                    print('keep dsnb greater than atm  method2 DT = ', DT_method2_good3, '\n')
                    
                    #keep dsnb greater than else bgs and greater than fraction of atm
                    DT_method2_good4,_,_,_,_ = get_nume_DT_alphabeta_G_method2(exposures, 
                                        null4_sum_ijkappa.value, alter4_sum_ijkappa.value, 
                                        alpha,power, simnum = 500000, 
                                         plot= False, print_check = False)
                    result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_lessatmlesselseMethod2'][str(alpha_power)] = DT_method2_good4

                    if end_index1-start_index1>=30:
                        DT_method1_good1 = get_analytic_calDcalT_shape(pd.DataFrame(rates_null_good1), 
                                                               pd.DataFrame(rates_det_good1), 
                                                               Phi_minus1, Z_alpha)
                        print('keep all dsnb method1 DT = ', DT_method1_good1, '\n')
                        result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_allDSNBMethod1'][str(alpha_power)] = DT_method1_good1
                    if end_index2-start_index2>=30:
                        DT_method1_good2 = get_analytic_calDcalT_shape(pd.DataFrame(rates_null_good2), 
                                                           pd.DataFrame(rates_det_good2), 
                                                           Phi_minus1, Z_alpha)
                        print('remove all bgs except atm and dsnb exists method1 DT = ', DT_method1_good2, '\n')
                        result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_noelsebgMethod1'][str(alpha_power)] = DT_method1_good2

                    if end_index3-start_index3>=30:
                        DT_method1_good3 = get_analytic_calDcalT_shape(pd.DataFrame(rates_null_good3), 
                                                           pd.DataFrame(rates_det_good3), 
                                                           Phi_minus1, Z_alpha)
                        print('keep dsnb greater than atm method1 DT = ', DT_method1_good3, '\n')
                        result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_lessatmbgMethod1'][str(alpha_power)] = DT_method1_good3

                    if end_index4-start_index4>=30:
                        DT_method1_good4 = get_analytic_calDcalT_shape(pd.DataFrame(rates_null_good4), 
                                                           pd.DataFrame(rates_det_good4), 
                                                           Phi_minus1, Z_alpha)
                        print('keep dsnb greater than else bgs and greater than fraction of atm method1 DT = ', DT_method1_good4, '\n')
                        result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_lessatmlesselseMethod1'][str(alpha_power)] = DT_method1_good4





```
