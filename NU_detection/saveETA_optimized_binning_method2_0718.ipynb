```python
folder_2d = '../nestpy_unbinned_unbound_thrd0'
E_threshold_keV = 0
```

```python
signal_pcles  = ['atm', 'hep', 'dsnb']
```

```python
detectors = ['CJPL', 'Kamioka', 'LNGS', 'SURF', 'SNOlab']
metallicities = 'high', 'low'
```

```python
etaijk_pd = pd.DataFrame(
                         index =[signal_pcle + '_'+detector+'_'+metallicity for signal_pcle in signal_pcles 
                                 for detector in detectors 
                                  for  metallicity in metallicities],
                        columns = ['null_sum', 'alter_sum'])
```

```python
etaijk_pd
```

```python
d = dict()
for signal_pcle in signal_pcles:
    file_name = 'optimized_consecutive_binning_thrd0_'+signal_pcle+'.csv'

    for detector in detectors:

        optimized_bin_dets = pd.read_csv(file_name, index_col = 0)
        print(file_name)
        
        
        if signal_pcle == 'dsnb':
            
            pcles_det = ['dsnb']
            pcles_null, _, pcles_det, _, _ = get_null_NR(signal_pcle, detector)
            bins_groups_det = [literal_eval(a) for a in 
                           [subbin for subbin in list(optimized_bin_dets['dsnb']) if str(subbin)!='nan']]
        elif signal_pcle == 'hep':
            
            pcles_det = ['hep']
            pcles_null, _, pcles_det, _, _ = get_null_NR(signal_pcle, detector)
            bins_groups_det = [literal_eval(a) for a in 
                           [subbin for subbin in list(optimized_bin_dets['hep']) if str(subbin)!='nan']]
        elif signal_pcle == 'atm':
            
            pcles_det = ['atmNu_'+detector+'_avg']
            pcles_null, pcles_det, _, _ = get_null_NR(signal_pcle, detector)
            bins_groups_det = [literal_eval(a) for a in 
                           [subbin for subbin in list(optimized_bin_dets[detector]) if str(subbin)!='nan']]
        
        print(pcles_null)

        for m, metallicity in enumerate(metallicities):
            _, sim_rate_null = get_poisrdm_Rate(pcles_null, bins_groups_det, N = 1, 
                         nuclei_2d = 'Xenon', footnote = 'thrd', E_threshold_keV = E_threshold_keV, metallicity =metallicity,
                         folder_2d = folder_2d,plot = False, print_check = False, 
                                                   plot_xlims = [0, 150], plot_ylims = [2, 5])

            _, sim_rate_det = get_poisrdm_Rate(pcles_det, bins_groups_det, N = 1, 
                             nuclei_2d = 'Xenon', footnote = 'thrd', E_threshold_keV = E_threshold_keV, metallicity = metallicity,
                             folder_2d = folder_2d,plot = True, print_check = False, 
                                                       plot_xlims = [0, 150], plot_ylims = [2, 5])


            null_sum_ijkappa = sum(sim_rate_null).to(1/u.tonne/u.yr).value
            alter_sum_ijkappa = sum(sim_rate_det+sim_rate_null).to(1/u.tonne/u.yr).value
            print(detector, metallicity, 'null_sum_ij = ', null_sum_ijkappa, 
                  '\nalter_sum_ij = ', alter_sum_ijkappa)

            d[signal_pcle + '_'+detector+'_'+metallicity + '_null_i'] = (sim_rate_null).to(1/u.tonne/u.yr).value
            d[signal_pcle + '_'+detector+'_'+metallicity + '_alter_i'] = (sim_rate_det+sim_rate_null).to(1/u.tonne/u.yr).value
            
            etaijk_pd['null_sum'][signal_pcle + '_'+detector+'_'+metallicity] = null_sum_ijkappa
            etaijk_pd['alter_sum'][signal_pcle + '_'+detector+'_'+metallicity] = alter_sum_ijkappa

         
```

```python
binned_etaijk_pd = pd.DataFrame.from_dict(d,orient='index').transpose()
binned_etaijk_pd
```
