```python


alphas_powers = [(0.001, 0.9), (0.001, 0.8), (0.01, 0.9), (0.01, 0.8),(0.1, 0.9), (0.1, 0.8)]

result_pd = pd.DataFrame(index = [str(alpha_power) for alpha_power in alphas_powers],
                         columns=[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_'+method for detector in detectors 
                                  for  metallicity in metallicities
                                  for eff_corr in effs_corrs 
                                 for method in ['allHEPMethod1', 'allHEPMethod2', 
                                                'no8BgMethod1', 'no8BbgMethod2']])
```

```python

exposures = [int(i) for i in np.logspace(np.log10(100), np.log10(6000), 26)]
Er_bincenters = list(binned_1d_pd['Xenon_Er_bincenters [keV]'])
for eff_corr in effs_corrs:
    eff, corr = eff_corr 
    print(eff_corr)
    for detector in detectors:
        print(detector)
        for metallicity in metallicities:
            pcles_det = ['hep']
            select_col_det  = [col for col in binned_1d_pd.columns 
                               if '_'+eff+'_' in col and corr in col and 
                               metallicity in col and any(pcle in col for pcle in pcles_det)]

            pcles_null_NR_wo8B =['dsnb','atmNu_'+detector+'_avg']
            pcles_null_NR = ['dsnb', '8B','atmNu_'+detector+'_avg']
            pcles_null_NR_8B = ['8B']
          
            select_col_null_wo8B  = [col for col in binned_1d_pd.columns 
                                      if '_'+eff+'_' in col and corr in col and 
                                      metallicity in col and any(pcle in col for pcle in pcles_null_NR_wo8B)]
            select_col_null  = [col for col in binned_1d_pd.columns 
                                if '_'+eff+'_' in col and corr in col and 
                                metallicity in col and any(pcle in col for pcle in pcles_null_NR)]
            select_col_null_8B  = [col for col in binned_1d_pd.columns 
                                if '_'+eff+'_' in col and corr in col and 
                                metallicity in col and any(pcle in col for pcle in pcles_null_NR_8B)]

            if len(
                select_col_det) !=len(pcles_det) or len(
                select_col_null) !=len(pcles_null_NR) or len(
                select_col_null_wo8B) !=len(pcles_null_NR_wo8B)or len(
                select_col_null_8B) !=len(pcles_null_NR_8B):
                print('wrong selection',
                     select_col_det, pcles_det,'\n',
                     select_col_null, pcles_null_NR, '\n',
                     select_col_null_wo8B, pcles_null_NR_wo8B, '\n',
                     select_col_null_8B, pcles_null_NR_8B)
                break
            else:

                rates_det = list(np.sum(binned_1d_pd[select_col_det], axis = 1))/u.tonne/u.yr
                rates_null = list(np.sum(binned_1d_pd[select_col_null], axis = 1))/u.tonne/u.yr
                rates_null_wo8B = list(np.sum(binned_1d_pd[select_col_null_wo8B], axis = 1))/u.tonne/u.yr
                rates_null_8B = list(np.sum(binned_1d_pd[select_col_null_8B], axis = 1))/u.tonne/u.yr
               

                #keep all hep
                good1 = rates_det>0

                #remove all 8B and hep exists
                good2 = np.multiply(rates_det>0, rates_null_8B==0)

             
                start_index1, end_index1 = min(np.where(good1)[0]), max(np.where(good1)[0])
                start_index2, end_index2 = min(np.where(good2)[0]), max(np.where(good2)[0])
             
                rates_null_good1 = rates_null[start_index1:end_index1+1]
                rates_null_good2 = rates_null[start_index2:end_index2+1]
               
                rates_det_good1 = rates_det[start_index1:end_index1+1]
                rates_det_good2 = rates_det[start_index2:end_index2+1]
               
                null1_sum_ijkappa = sum(rates_null_good1)
                alter1_sum_ijkappa = sum(rates_det_good1 + rates_null_good1)

                null2_sum_ijkappa = sum(rates_null_good2)
                alter2_sum_ijkappa = sum(rates_det_good2 + rates_null_good2)

              
                print('\nEr range keep all hep: ', len(rates_null_good1), 'bins from ', 
                      Er_bincenters[start_index1], ' to ', Er_bincenters[end_index1], 
                      '\nnull_sum_ijkappa = ', null1_sum_ijkappa, 
                      '\nalter_sum_ijkappa = ', alter1_sum_ijkappa,

                     '\n\nEr range remove all 8B and hep exists: ', len(rates_null_good2), 'bins from ', 
                      Er_bincenters[start_index2], ' to ', Er_bincenters[end_index2], 
                      '\nnull_sum_ijkappa = ', null2_sum_ijkappa, 
                      '\nalter_sum_ijkappa = ', alter2_sum_ijkappa, 

                      '\n')

                for alpha_power in alphas_powers:
                    dict_key = str(alpha_power)
                    alpha, power = alpha_power
                    Z_alpha = round(stats.norm.ppf(alpha), 3)
                    Phi_minus1 = round(stats.norm.ppf(power), 3)


                    #keep all hep
                    DT_method2_good1,_,_,_,_ = get_nume_DT_alphabeta_G_method2(exposures, 
                                        null1_sum_ijkappa.value, alter1_sum_ijkappa.value, 
                                        alpha, power, simnum = 500000, 
                                         plot= False, print_check = False)
                    result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_allHEPMethod2'][str(alpha_power)] = DT_method2_good1
                    print('keep all hep method2 DT = ', DT_method2_good1, '\n')

                    #remove all 8B and hep exists
                    DT_method2_good2,_,_,_,_ = get_nume_DT_alphabeta_G_method2(exposures, 
                                        null2_sum_ijkappa.value, alter2_sum_ijkappa.value, 
                                        alpha, power, simnum = 500000, 
                                         plot= False, print_check = False)
                    result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_no8BbgMethod2'][str(alpha_power)] = DT_method2_good2
                    print('remove all 8B and hep exists method2 DT = ', DT_method2_good2, '\n')

                
                    if end_index1-start_index1>=30:
                        DT_method1_good1 = get_analytic_calDcalT_shape(pd.DataFrame(rates_null_good1), 
                                                               pd.DataFrame(rates_det_good1), 
                                                               Phi_minus1, Z_alpha)
                        print('keep all hep method1 DT = ', DT_method1_good1, '\n')
                        result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_allHEPMethod1'][str(alpha_power)] = DT_method1_good1
                    if end_index2-start_index2>=30:
                        DT_method1_good2 = get_analytic_calDcalT_shape(pd.DataFrame(rates_null_good2), 
                                                           pd.DataFrame(rates_det_good2), 
                                                           Phi_minus1, Z_alpha)
                        print('remove all 8B and hep exists method1 DT = ', DT_method1_good2, '\n')
                        result_pd[detector+'_'+metallicity+'_'+(''.join(eff_corr))+'_no8BbgMethod1'][str(alpha_power)] = DT_method1_good2

              
```
